<br>**Link:** https://shopee.co.id/api/v4/search/search_items?by=relevancy&keyword=iphone%2012&limit=60&newest=0&order=desc&page_type=search&scenario=PAGE_GLOBAL_SEARCH&version=2 <br>
<br> **Reference:** https://youtu.be/J7PZoVoP2Ng

```python
shopee_url = 'https://shopee.co.id'
keyword = 'Iphone 12 Pro'

header = {
    'User-Agent' : 'Chrome',
    'Referer' : '{}search?keyword={}'.format(shopee_url,keyword)
}

url = 'https://shopee.co.id/api/v4/search/search_items?by=relevancy&keyword={}&limit=60&newest=0&order=desc&page_type=search&scenario=PAGE_GLOBAL_SEARCH&version=2'.format(keyword) #limitnya bisa di custom

#API Request
r = requests.get(url,headers = header).json()

#Scraping
col_list,price_list,sold_list,rating_list,location_list = [],[],[],[],[]
for item in r['items']:
  col_list.append(item['item_basic']['name'])
  price_list.append(item['item_basic']['price'])
  sold_list.append(item['item_basic']['sold'])
  rating_list.append(item['item_basic']['item_rating']['rating_star'])
  location_list.append(item['item_basic']['shop_location'])
```

```python
# print the data
print(col_list)

# count the data
len(col_list)
```

```python
#Save data
listCols=['col_list','price_list','sold_list','rating_list','location_list']
dict_data = dict(zip(
    listCols,
    (col_list,price_list,sold_list,rating_list,location_list)
))
```

```python
import json
with open('digimao.json','w') as fp:
  json.dump(dict_data,fp)
```
